```python
# Task: Load and Explore the Dataset (Web-based Jupyter Notebook Version)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Enable inline plotting for Jupyter environments
%matplotlib inline

# Load Iris dataset using sklearn
iris = load_iris(as_frame=True)
df = iris.frame

# Clean column names for clarity
df.columns = [col.replace(' (cm)', '').replace(' ', '_') for col in df.columns]

# Display the first few rows
print("\nFirst 5 rows:")
display(df.head())

# Display info
print("\nDataset Info:")
df.info()

# Check for missing values
print("\nMissing Values:")
display(df.isnull().sum())

# Clean data by dropping rows with missing values
df = df.dropna()

# Task 2: Basic Data Analysis
print("\nDescriptive Statistics:")
display(df.describe())

# Grouping by species and computing mean
if 'target' in df.columns:
    print("\nMean values per species:")
    display(df.groupby('target').mean())

# Task 3: Data Visualization
# Line chart - sepal_length over index
if 'sepal_length' in df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df['sepal_length'], label='Sepal Length')
    plt.title('Sepal Length Over Index')
    plt.xlabel('Index')
    plt.ylabel('Sepal Length')
    plt.legend()
    plt.grid(True)
    plt.show()

# Bar chart - average petal_length per species
if 'target' in df.columns and 'petal_length' in df.columns:
    plt.figure(figsize=(8, 5))
    df.groupby('target')['petal_length'].mean().plot(kind='bar', color='coral', edgecolor='black')
    plt.title('Average Petal Length per Species')
    plt.xlabel('Species (target label)')
    plt.ylabel('Petal Length')
    plt.grid(axis='y')
    plt.show()

# Histogram of sepal_width
if 'sepal_width' in df.columns:
    plt.figure(figsize=(8, 5))
    plt.hist(df['sepal_width'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Distribution of Sepal Width')
    plt.xlabel('Sepal Width')
    plt.ylabel('Frequency')
    plt.grid(axis='y')
    plt.show()

# Scatter plot: sepal_length vs. petal_length
if 'sepal_length' in df.columns and 'petal_length' in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df['sepal_length'], df['petal_length'], c='purple', alpha=0.6)
    plt.title('Sepal Length vs. Petal Length')
    plt.xlabel('Sepal Length')
    plt.ylabel('Petal Length')
    plt.grid(True)
    plt.show()
```
